```python
# Dependencies
import csv
import numpy as np
from citipy import citipy
import random
import matplotlib.pyplot as plt
import requests as req
import pandas as pd
import json
import seaborn as sbn
```

# Data Analysis
Observe Trend 1: Temperature in Cities that are between latitudes of -20 and 20 are higher in general
Observe Trend 2: Lowest temperature at latitudes of 40 to 60
Observe Trend 3: Highest wind speed found between latitudes of 38 to 40


```python
#Randomly select latitude and longitude
randCity = []
randCountry = []
count = 0

##Loop through all coordinates to find cities using citipy

while len(randCity) < 600:
    count +=1
    ### Generate latitude
    lat_negpos = random.randint(-1,1)
    lat = lat_negpos*(random.randint(0,90)+random.random())
    
    ### Generate longitude
    lon_negpos = random.randint(-1,1)
    lon = lon_negpos*(random.randint(0,180)+random.random()/100)
    
    ### Find nearest city
    city = citipy.nearest_city(lat,lon)
    
    ### Put in the city and country list and ensure no repeats
    repeat = False
    for i in range(len(randCity)):
        if city.city_name == randCity[i]:
            repeat = True
    if repeat == False:
        randCity.append(city.city_name)
        randCountry.append(city.country_code)
```

```python
# Save config information.
api_key = "0914b51805c6b59471b88039b28e5a33"
endpoint = "http://api.openweathermap.org/data/2.5/weather"
units = "imperial"

# Build partial query URL
params = {'appid': api_key,'units': units}
```

```python
# Loop through the list of cities and perform a request for data on each
weather_data = []
cities = []
countries = []
counta=0
print("Beginning Data Retrieval")
print("--------------------------------------------------------")
for city in randCity:
    counta +=1
    print("Processing Record " + str(counta) + " | " + city)
    params['q'] = city
    response = req.get(endpoint, params=params).json()
    cityName=response.get("name")
    #Only include cities that have weather data available
    if response.get("name"):
        weather_data.append(response)
        cities.append(city)
        countries.append(randCountry[counta-1])
        print("           Weather data found for " + cityName)
    else:
        print("           No weather data found")
    

print("--------------------------------------------------------")    
print("Data Retrieval Complete")    
print("--------------------------------------------------------")
```

```python
# Extract Temperature (F), Humidity, Cloudiness, Wind Speed, and Latitude data from responses
temp_data = [data.get("main").get("temp") for data in weather_data]
humid_data = [data.get("main").get("humidity") for data in weather_data]
cloud_data = [data.get("clouds").get("all") for data in weather_data]
wind_data = [data.get("wind").get("speed") for data in weather_data]
lat_data = [data.get("coord").get("lat") for data in weather_data]

weather = {"city":cities, "country":countries, 
           "temp":temp_data, "humidity":humid_data, "cloud":cloud_data, "wind":wind_data, "latitude":lat_data}
weather_df = pd.DataFrame(weather)
weather_df.head()
```

```python
#Save weather DataFrame as csv
weather_df.to_csv("weatherdata.csv",index=False)
```

```python
# Build a scatter plot for Temperature vs. Latitude
plt.scatter(weather_df["latitude"], weather_df["temp"], marker="o", color ="blue", edgecolors ="green")

# Incorporate the other graph properties
plt.title("Temperature in World Cities")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim(-90,90)

# Save the figure
plt.savefig("TemperatureInWorldCities.png")

# Show plot
plt.show()
```
